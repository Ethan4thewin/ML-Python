```python
import mglearn
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split #split the data into 75 train / 25 test, and random the order
```

```python
x,y = mglearn.datasets.make_forge()
```

```python
# Splitting
x_train,x_test = train_test_split(x, random_state=0)
y_train,y_test = train_test_split(y, random_state=0)
```

```python
# import KNeighbor algorithm
from sklearn.neighbors import KNeighborsClassifier
# Define the algorithm behavior with 3 closest neighbors
clf = KNeighborsClassifier(n_neighbors=3)
# fit the classifier using the training set (storing the dataset for prediction)
clf.fit(x_train,y_train)
```

```python
# 1st solution: storing prediction and compare the pred with the test labels to evaluate how well our model generalizes
y_pred = clf.predict(x_test)
np.mean(y_pred == y_test)
```

```python
# 2nd solution: call the test data with the test label together
clf.score(x_test,y_test)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 3))
for n_neighbors, ax in zip([1, 3, 9], axes):
    # the fit method returns the object self, so we can instantiate
    # and fit in one line
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(x, y)
    mglearn.plots.plot_2d_separator(clf, x, fill=True, eps=.5, ax=ax, alpha=.3)
    mglearn.discrete_scatter(x[:, 0], x[:, 1], y, ax=ax)
    ax.set_title("{} neighbor(s)".format(n_neighbors))
    ax.set_xlabel("feature 0")
    ax.set_ylabel("feature 1")
axes[0].legend(loc=3)
```
