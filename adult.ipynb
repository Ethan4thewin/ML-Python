```python
import pandas as pd
data = pd.read_csv("https://raw.githubusercontent.com/amueller/introduction_to_ml_with_python/master/data/adult.data",
    header=None, index_col=False,
    names=['age', 'workclass', 'fnlwgt', 'education', 'education-num',
'marital-status', 'occupation', 'relationship', 'race', 'gender',
'capital-gain', 'capital-loss', 'hours-per-week', 'native-country',
'income']
)
```

```python
head = ['age', 'workclass', 'education', 'gender', 'hours-per-week', 'occupation', 'income']
short_data = data[head]
short_data.head()
```

```python
short_data_dummies = pd.get_dummies(short_data)
short_data_dummies
```

```python
features = short_data_dummies.iloc[:,0:44]
x = features.values
y = short_data_dummies['income_ >50K'].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
logreg = LogisticRegression(max_iter=9999).fit(x_train, y_train)
logreg.score(x_test, y_test)
```
